```python
#Use Hodrick-Prescott Filter to find trend and noise for housing prices from 2011-2021
import statsmodels.api as sm

collin_noise, collin_trend = sm.tsa.filters.hpfilter(collin_df['Average Price'])
dallas_noise, dallas_trend = sm.tsa.filters.hpfilter(dallas_df['Average Price'])
denton_noise, denton_trend = sm.tsa.filters.hpfilter(denton_df['Average Price'])
ellis_noise, ellis_trend = sm.tsa.filters.hpfilter(ellis_df['Average Price'])
johnson_noise, johnson_trend = sm.tsa.filters.hpfilter(johnson_df['Average Price'])
kaufman_noise, kaufman_trend = sm.tsa.filters.hpfilter(kaufman_df['Average Price'])
parker_noise, parker_trend = sm.tsa.filters.hpfilter(parker_df['Average Price'])
rockwall_noise, rockwall_trend = sm.tsa.filters.hpfilter(rockwall_df['Average Price'])
tarrant_noise, tarrant_trend = sm.tsa.filters.hpfilter(tarrant_df['Average Price'])
```
