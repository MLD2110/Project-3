```python
#Create a Linear Regression model and fit it to the training data
from sklearn.linear_model import LinearRegression

collin_model = LinearRegression()
collin_model.fit(collin_X_train, collin_y_train)

dallas_model = LinearRegression()
dallas_model.fit(dallas_X_train, dallas_y_train)

denton_model = LinearRegression()
denton_model.fit(denton_X_train, denton_y_train)

ellis_model = LinearRegression()
ellis_model.fit(ellis_X_train, ellis_y_train)

johnson_model = LinearRegression()
johnson_model.fit(johnson_X_train, johnson_y_train)

kaufman_model = LinearRegression()
kaufman_model.fit(kaufman_X_train, kaufman_y_train)

parker_model = LinearRegression()
parker_model.fit(parker_X_train, parker_y_train)

rockwall_model = LinearRegression()
rockwall_model.fit(rockwall_X_train, rockwall_y_train)

tarrant_model = LinearRegression()
tarrant_model.fit(tarrant_X_train, tarrant_y_train)
```
